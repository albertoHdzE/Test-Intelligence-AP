````python
OIA_ds = pd.read_csv("data/data.csv")
arnold_df = copy.deepcopy(OIA_ds[OIA_ds["language"] == "ArnoldC"])
original_sequences = arnold_df["sequence"].values

arnoldC_code_list = (OIA_ds[OIA_ds["language"] == "ArnoldC"])["GPT4_answer"].values
total_questions = len(arnoldC_code_list)
````

````python
counter = -1
exe_outputs = []
levenshtein_list = []
correctness_list = []
test_code = []

for code in arnoldC_code_list:#[0:15]:
    code = code.replace("```arnoldc" + "\n","\"")
    code = code.replace("```ArnoldC" + "\n","\"")
    code = code.replace("\n"+ "```","\"")
    
    
    #print(code)
    counter += 1
    name_file = "q_" + str(counter)
    surname_file =  '.arnoldc'
    whole_name = name_file + surname_file
    
    original_seq = original_sequences[counter]
    
    print("######################  INITIALIZE PROCESS ########################## (", str(counter), ")")
    try:
        print("creating file...")
        test_code.append(code)
        ! echo -e {code} > {whole_name}
        #COMPLIE
        print("----------------- compliling")
        rough_jar = subprocess.run(["java", "-jar",  "ArnoldC.jar", whole_name], capture_output=True, text=True)
        print("SUCCESSFUL COMPILATION")
        print(rough_jar.stdout.strip())

        try:
            print("-----------------  executing")
            rough_jar = subprocess.run(["java", name_file], capture_output=True, text=True, timeout=10)
            print("EXECUTION OUT: ")
            print(rough_jar.stdout.strip())
            out = rough_jar.stdout.strip()
            out = out.replace("\n", ", ")
            exe_outputs.append(out)
            correctness_list.append(out==original_seq)
            levenshtein_list.append(levenshtein(out, original_seq))
        except:
            print("**************** execution error ( step 2)")
            print(rough_jar.stderr)
            exe_outputs.append('Execution error')
            correctness_list.append('False')
            levenshtein_list.append(50000)
            pass

    except:       
        print("**************** Execution error (step 1)")
        print(rough_jar.stderr)
        pass
    
    
  
    
````

````python
import pandas as pd
import subprocess
import time
import copy
from enchant.utils import levenshtein
````

````python
OIA_ds = pd.read_csv("data/data.csv")
m_df = copy.deepcopy(OIA_ds[OIA_ds["language"] == "Mathematica"])
original_sequences = m_df["sequence"].values

m_code_list = (OIA_ds[OIA_ds["language"] == "Mathematica"])["GPT4_answer"].values
total_questions = len(m_code_list)
````

````python
m_df
````

````python
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr

session = WolframLanguageSession()
````

````python
counter = -1
exe_outputs = []
levenshtein_list = []
correctness_list = []
test_code = []

for code in m_code_list:  # [0:15]:
    code = code.replace("```mathematica" + "\n", "")
    code = code.replace("```Mathematica" + "\n", "")
    code = code.replace("\n" + "```", "")

    # print(code)
    counter += 1
    name_file = "m_" + str(counter)
    surname_file = ".m"
    whole_name = name_file + surname_file

    original_seq = original_sequences[counter]

    print(
        "######################  INITIALIZE PROCESS ########################## (",
        str(counter),
        ")",
    )
    try:
        # print(code)
        out = str(session.evaluate(wlexpr(code)))

        # print("creating file...")
        # test_code.append(code)
        # ! echo -e {code} > {whole_name}

        # rough_jar = subprocess.run(["session.evaluate(wlexpr('{}')".format(code)], capture_output=True, text=True, timeout=10)
        # print("EXECUTION OUT: ")
        # # print(rough_jar.stdout.strip())
        # out = rough_jar.stdout.strip()
        out = out.replace("PackedArray([ ", "")
        out = out.replace("], dtype=int8),", "")
        out = out.replace("(", "")
        out = out.replace(")", "")
        out = out.replace("[ ", "")
        out = out.replace("  ", ", ")
        out = out.replace(" ", ", ")
        out = out.replace("]", "")
        out = out.replace("[", "")
        out = out.replace("\n", ", ")
        out = out.replace("FindSequenceFunction", "")
        out = out.replace(",  ", ", ")
        out = out.replace(", , ", ", ")

        fist = out[:2]
        if fist == ", ":
            out = out[2 : len(out) - 1]
        fist = out[:3]
        if fist == ",, ":
            out = out[3 : len(out) - 1]
        out = out.replace(",, ", ", ")
        out = out.replace("BaseForm", "")
        out = out.replace("DirectedInfinity", "")
        out = out.replace("nan,, ", "")
        out = out.replace("TotalTake", "")
        out = out.replace("', '", ", ")
        out = out.replace("Plus", "")
        out = out.replace("Part", "")
        out = out.replace("None", "")
        out = out.replace(", , ", ", ")
        # out = out.replace(",, TotalTakeTotalTakeTotalTake", ", ")
        # out = out(",, TotalTakeTotalTake", ", ")
        print(out)
        exe_outputs.append(out)
        correctness_list.append(out == original_seq)
        levenshtein_list.append(levenshtein(out, original_seq))

    except:
        print("**************** execution error ( step 2)")
        print(rough_jar.stderr)
        exe_outputs.append("Execution error")
        correctness_list.append("False")
        levenshtein_list.append(50000)
        pass
````
