```python
import os

# Import writer class from csv module
from csv import writer


def append_result_csv(results_list: list, out_file_name: str):
    # Open our existing CSV file in append mode
    # Create a file object for this file
    if not os.path.isfile(out_file_name):
        with open(out_file_name, "a") as f_object:
            writer_object = writer(f_object)
            # writer_object.writerow(["complexity", "length_type", "language", "q_length", "a_length", "temperature", "print_code", "real_code", "incapacity"])
            writer_object.writerow(
                [
                    "sequence",
                    "complexity",
                    "length_type",
                    "language",
                    "q_length",
                    "answer",
                    "a_length",
                    "temperature",
                    "real_code",
                    "print_code",
                    "no_code",
                    "explanation",
                    "no_sense",
                ]
            )
            writer_object.writerow(results_list)
    else:
        with open(out_file_name, "a") as f_object:
            # Pass this file object to csv.writer()
            # and get a writer object
            writer_object = writer(f_object)
            # Pass the list as an argument into
            # the writerow()
            writer_object.writerow(results_list)
            # Close the file object
            f_object.close()
```

```python
def get_chatGPT_4_answers(
    questions_bunch, complexity, length_type, language, temperature, csv_name
):

    for q in questions_bunch:
        index = q.index("sequence")
        seq_str = (
            "Complexity: " + str(complexity) + "[" + q[index + 8 : len(q) - 1] + "]"
        )

        try:
            response = openai.ChatCompletion.create(
                model="gpt-4-turbo-preview",  # "gpt-4-turbo-preview",
                messages=[
                    {"role": "system", "content": "You are a helpful assistant."},
                    {"role": "user", "content": q},
                ],
                temperature=temperature,
                request_timeout=2500,
            )

            text = response.choices[0].message["content"]

            q_length = len(q)
            a_length = len(text)
            results = [
                seq_str,
                complexity,
                length_type,
                language,
                q_length,
                text,
                a_length,
                temperature,
                0,
                0,
                0,
                0,
                0,
            ]
            append_result_csv(results, csv_name)

            print("*** ", seq_str, "***")
            print("R: ", text)
        except OpenAIError as e:
            # Handle all OpenAI API errors
            results = [
                seq_str,
                complexity,
                length_type,
                language,
                q_length,
                "ERROR!!!",
                a_length,
                temperature,
                0,
                0,
                0,
                0,
                0,
            ]
            append_result_csv(results, csv_name)
            print("*******************  ERROR!!!****************")
            print(f"Error: {e}")
```
